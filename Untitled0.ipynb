````python
class LayerNorm(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.w = nn.Parameter(torch.ones(cfg.d_model))
        self.b = nn.Parameter(torch.zeros(cfg.d_model))
        # Define the dimension over which to normalize
        self.normalized_shape = cfg.d_model

    def forward(self, residual):
        # residual: [batch, position, d_model]
        # Calculate mean and variance along the last dimension
        mean = residual.mean(dim=-1, keepdim=True)
        var = residual.var(dim=-1, keepdim=True, unbiased=False) # Use unbiased=False for consistency with PyTorch LayerNorm

        # Normalize the input
        normalized_residual = (residual - mean) / torch.sqrt(var + self.cfg.layer_norm_eps)

        # Scale and shift with learned parameters
        output = normalized_residual * self.w + self.b
        return output
_ = rand_float_test(LayerNorm, [2, 4, 768])
_ = load_gpt2_test(LayerNorm, reference_gpt2.ln_final, "blocks.11.hook_resid_post")
````

````python
class Attention(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.W_Q = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_model, cfg.d_head)))
        nn.init.normal_(self.W_Q, std=self.cfg.init_range)
        self.b_Q = nn.Parameter(torch.zeros((cfg.n_heads, cfg.d_head)))
        self.W_K = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_model, cfg.d_head)))
        nn.init.normal_(self.W_K, std=self.cfg.init_range)
        self.b_K = nn.Parameter(torch.zeros((cfg.n_heads, cfg.d_head)))
        self.W_V = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_model, cfg.d_head)))
        nn.init.normal_(self.W_V, std=self.cfg.init_range)
        self.b_V = nn.Parameter(torch.zeros((cfg.n_heads, cfg.d_head)))

        self.W_O = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_head, cfg.d_model)))
        nn.init.normal_(self.W_O, std=self.cfg.init_range)
        self.b_O = nn.Parameter(torch.zeros((cfg.d_model)))

        self.register_buffer("IGNORE", torch.tensor(-1e5, dtype=torch.float32, device="cuda"))

    def forward(self, normalized_resid_pre):
        # Calculate Q, K, V
        q = einsum('batch pos d_model, n_heads d_model d_head -> batch pos n_heads d_head',
                  normalized_resid_pre, self.W_Q) + self.b_Q
        k = einsum('batch pos d_model, n_heads d_model d_head -> batch pos n_heads d_head',
                  normalized_resid_pre, self.W_K) + self.b_K
        v = einsum('batch pos d_model, n_heads d_model d_head -> batch pos n_heads d_head',
                  normalized_resid_pre, self.W_V) + self.b_V

        # Calculate attention scores
        attn_scores = einsum('batch q_pos n_heads d_head, batch k_pos n_heads d_head -> batch n_heads q_pos k_pos',
                           q, k) / math.sqrt(self.cfg.d_head)

        # Apply causal mask
        mask = torch.triu(torch.ones(attn_scores.size(-2), attn_scores.size(-1),
                                   device=attn_scores.device), diagonal=1).bool()
        attn_scores = attn_scores.masked_fill(mask, self.IGNORE)

        # Softmax and attention output
        attn_weights = attn_scores.softmax(dim=-1)
        z = einsum('batch n_heads q_pos k_pos, batch k_pos n_heads d_head -> batch q_pos n_heads d_head',
                  attn_weights, v)

        # Combine heads
        out = einsum('batch pos n_heads d_head, n_heads d_head d_model -> batch pos d_model',
                    z, self.W_O) + self.b_O
        return out
````

## MLP

````python
class MLP(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.W_in = nn.Parameter(torch.empty((cfg.d_model, cfg.d_mlp)))
        nn.init.normal_(self.W_in, std=self.cfg.init_range)
        self.b_in = nn.Parameter(torch.zeros((cfg.d_mlp)))
        self.W_out = nn.Parameter(torch.empty((cfg.d_mlp, cfg.d_model)))
        nn.init.normal_(self.W_out, std=self.cfg.init_range)
        self.b_out = nn.Parameter(torch.zeros((cfg.d_model)))

    def forward(self, normalized_resid_mid):
        # normalized_resid_mid: [batch, position, d_model]
        # YOUR CODE HERE - IMPLEMENT THE MLP FORWARD PASS
        # Step 1: Linear transformation with W_in and b_in
        hidden = einops.einsum(normalized_resid_mid, self.W_in, 'batch position d_model, d_model d_mlp -> batch position d_mlp') + self.b_in
        # Step 2: Apply GELU activation
        activated = gelu_new(hidden)
        # Step 3: Linear transformation with W_out and b_out
        output = einops.einsum(activated, self.W_out, 'batch position d_mlp, d_mlp d_model -> batch position d_model') + self.b_out
        return output

# The rand_float_test and load_gpt2_test calls remain the same and should now work.
# rand_float_test(MLP, [2, 4, 768])
# load_gpt2_test(MLP, reference_gpt2.blocks[0].mlp, cache["blocks.0.ln2.hook_normalized"])
````

## Transformer Block

````python
class TransformerBlock(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg

        self.ln1 = LayerNorm(cfg)
        self.attn = Attention(cfg)
        self.ln2 = LayerNorm(cfg)
        self.mlp = MLP(cfg)

    def forward(self, resid_pre):
        # resid_pre [batch, position, d_model]
        # Apply LayerNorm, Attention, and residual connection
        resid_mid = self.attn(self.ln1(resid_pre)) + resid_pre
        # Apply LayerNorm, MLP, and residual connection
        resid_post = self.mlp(self.ln2(resid_mid)) + resid_mid
        return resid_post

rand_float_test(TransformerBlock, [2, 4, 768])
load_gpt2_test(TransformerBlock, reference_gpt2.blocks[0], cache["resid_pre", 0])
````

## Unembedding

````python
class Unembed(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.W_U = nn.Parameter(torch.empty((cfg.d_model, cfg.d_vocab)))
        nn.init.normal_(self.W_U, std=self.cfg.init_range)
        self.b_U = nn.Parameter(torch.zeros((cfg.d_vocab), requires_grad=False))

    def forward(self, normalized_resid_final):
        # normalized_resid_final [batch, position, d_model]
        # YOUR CODE HERE - IMPLEMENT THE UNEMBEDDING FORWARD PASS
        # Perform a linear transformation: normalized_resid_final @ W_U + b_U
        # Use einsum for clarity and broadcasting the bias
        # Correct the order of arguments for einsum
        output = einsum('batch position d_model, d_model d_vocab -> batch position d_vocab', normalized_resid_final, self.W_U) + self.b_U
        return output

# The rand_float_test and load_gpt2_test calls remain the same and should now work.
rand_float_test(Unembed, [2, 4, 768])
load_gpt2_test(Unembed, reference_gpt2.unembed, cache["ln_final.hook_normalized"])
````

## Full Transformer

````python
class Embed(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.W_E = nn.Parameter(torch.empty((cfg.d_vocab, cfg.d_model)))
        nn.init.normal_(self.W_E, std=self.cfg.init_range)

    def forward(self, tokens):
        # tokens: [batch, position]
        # Perform embedding lookup.
        # The tokens tensor contains indices, and we use it to select rows from the W_E matrix.
        return self.W_E[tokens]

class PosEmbed(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        # W_pos is a lookup table for positional embeddings
        self.W_pos = nn.Parameter(torch.empty((cfg.n_ctx, cfg.d_model)))
        nn.init.normal_(self.W_pos, std=self.cfg.init_range)

    def forward(self, tokens):
        # tokens: [batch, position]
        # Create a tensor of positions [0, 1, 2, ..., position-1]
        position = torch.arange(tokens.size(1), device=tokens.device)
        # Lookup the positional embedding for each position in the batch
        # W_pos has shape [n_ctx, d_model]
        # We select the rows corresponding to the positions in the input sequence
        return self.W_pos[position]

class DemoTransformer(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.embed = Embed(cfg)
        self.pos_embed = PosEmbed(cfg)
        self.blocks = nn.ModuleList([TransformerBlock(cfg) for _ in range(cfg.n_layers)])
        self.ln_final = LayerNorm(cfg)
        self.unembed = Unembed(cfg)

    def forward(self, tokens):
        # tokens [batch, position]
        # YOUR CODE HERE - IMPLEMENT THE FORWARD PASS FOR THE FULL TRANSFORMER
        # Step 1: Get token embeddings
        embedded_tokens = self.embed(tokens)
        # Step 2: Get positional embeddings
        pos_embedded = self.pos_embed(tokens) # Assuming PosEmbed takes tokens to get position
        # Step 3: Combine embeddings (residual stream initially)
        residual = embedded_tokens + pos_embedded
        # Step 4: Pass through transformer blocks
        for block in self.blocks:
            residual = block(residual)
        # Step 5: Apply final layer norm
        normalized_residual = self.ln_final(residual)
        # Step 6: Apply unembedding
        logits = self.unembed(normalized_residual)

        return logits # Return the calculated logits

# The rand_int_test and load_gpt2_test calls remain the same and should now work.
rand_int_test(DemoTransformer, [2, 4])
load_gpt2_test(DemoTransformer, reference_gpt2, tokens)
````

````python
def lm_cross_entropy_loss(logits, tokens):
    # Measure next token loss
    # Logits have shape [batch, position, d_vocab]
    # Tokens have shape [batch, position]
    log_probs = logits.log_softmax(dim=-1)
    pred_log_probs = log_probs[:, :-1].gather(dim=-1, index=tokens[:, 1:].unsqueeze(-1)).squeeze(-1)
    return -pred_log_probs.mean()
loss = lm_cross_entropy_loss(demo_logits, test_tokens)
print(loss)
print("Loss as average prob", (-loss).exp())
print("Loss as 'uniform over this many variables'", (loss).exp())
print("Uniform loss over the vocab", math.log(demo_gpt2.cfg.d_vocab))
````

````python
class Attention(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.W_Q = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_model, cfg.d_head)))
        nn.init.normal_(self.W_Q, std=self.cfg.init_range)
        self.b_Q = nn.Parameter(torch.zeros((cfg.n_heads, cfg.d_head)))
        self.W_K = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_model, cfg.d_head)))
        nn.init.normal_(self.W_K, std=self.cfg.init_range)
        self.b_K = nn.Parameter(torch.zeros((cfg.n_heads, cfg.d_head)))
        self.W_V = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_model, cfg.d_head)))
        nn.init.normal_(self.W_V, std=self.cfg.init_range)
        self.b_V = nn.Parameter(torch.zeros((cfg.n_heads, cfg.d_head)))

        self.W_O = nn.Parameter(torch.empty((cfg.n_heads, cfg.d_head, cfg.d_model)))
        nn.init.normal_(self.W_O, std=self.cfg.init_range)
        self.b_O = nn.Parameter(torch.zeros((cfg.d_model)))

        self.register_buffer("IGNORE", torch.tensor(-1e5, dtype=torch.float32, device="cuda"))

    def forward(self, normalized_resid_pre):
        # normalized_resid_pre: [batch, position, d_model]

        # Calculate Query, Key, and Value matrices
        # W_Q: [n_heads, d_model, d_head], b_Q: [n_heads, d_head]
        # normalized_resid_pre: [batch, position, d_model]
        # Q, K, V: [batch, position, n_heads, d_head]
        Q = einsum('batch position d_model, n_heads d_model d_head -> batch position n_heads d_head', normalized_resid_pre, self.W_Q) + self.b_Q
        K = einsum('batch position d_model, n_heads d_model d_head -> batch position n_heads d_head', normalized_resid_pre, self.W_K) + self.b_K
        V = einsum('batch position d_model, n_heads d_model d_head -> batch position n_heads d_head', normalized_resid_pre, self.W_V) + self.b_V

        # Calculate attention scores
        # Q, K: [batch, position, n_heads, d_head]
        # attn_scores: [batch, n_heads, query_pos, key_pos]
        attn_scores = einsum('batch query_pos n_heads d_head, batch key_pos n_heads d_head -> batch n_heads query_pos key_pos', Q, K) / math.sqrt(self.cfg.d_head)

        # Apply causal mask
        # attn_scores: [batch, n_heads, query_pos, key_pos]
        attn_scores = self.apply_causal_mask(attn_scores)

        # Apply softmax to get attention probabilities
        # attn_probs: [batch, n_heads, query_pos, key_pos]
        attn_probs = attn_scores.softmax(dim=-1)

        # Calculate the output of the attention layer
        # attn_probs: [batch, n_heads, query_pos, key_pos]
        # V: [batch, key_pos, n_heads, d_head]
        # weighted_avg_values: [batch, query_pos, n_heads, d_head]
        weighted_avg_values = einsum('batch n_heads query_pos key_pos, batch key_pos n_heads d_head -> batch query_pos n_heads d_head', attn_probs, V)

        # Combine the heads
        # weighted_avg_values: [batch, position, n_heads, d_head] (query_pos is position here)
        # W_O: [n_heads, d_head, d_model]
        # output: [batch, position, d_model]
        output = einsum('batch position n_heads d_head, n_heads d_head d_model -> batch position d_model', weighted_avg_values, self.W_O) + self.b_O

        return output

    def apply_causal_mask(self, attn_scores):
        # attn_scores: [batch, n_heads, query_pos, key_pos]
        # Create a mask for the upper triangle of the key_pos dimension
        # The mask should be True where we want to mask (future tokens) and False otherwise
        query_pos = attn_scores.size(-2)
        key_pos = attn_scores.size(-1)
        # Create a lower triangular matrix
        mask = torch.triu(torch.ones(query_pos, key_pos, dtype=torch.bool, device=attn_scores.device), diagonal=1)
        # Expand mask to match the shape of attn_scores
        # mask: [query_pos, key_pos] -> [1, 1, query_pos, key_pos]
        attn_scores = attn_scores.masked_fill(mask, self.IGNORE)
        return attn_scores
````

````python
def train(model, train_loader, val_loader, optimizer, num_epochs, device='cuda'):
    model.train()
    best_val_loss = float('inf')

    for epoch in range(num_epochs):
        # Training phase
        model.train()
        train_loss = 0
        for batch in tqdm(train_loader, desc=f"Training Epoch {epoch+1}"):
            tokens = batch['tokens'].to(device)

            optimizer.zero_grad()
            logits = model(tokens)
            loss = lm_cross_entropy_loss(logits, tokens)
            loss.backward()

            # Gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

            optimizer.step()
            train_loss += loss.item()

        # Validation phase
        val_loss = evaluate(model, val_loader, device)

        print(f"Epoch {epoch+1}: Train Loss {train_loss/len(train_loader):.4f}, Val Loss {val_loss:.4f}")

        # Save best model
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), 'best_model.pt')
````

````python
def calculate_perplexity(model, data_loader, device='cuda'):
    model.eval()
    total_loss = 0
    total_tokens = 0

    with torch.no_grad():
        for batch in tqdm.tqdm(data_loader, desc="Evaluating"):
            tokens = batch['tokens'].to(device)
            logits = model(tokens)
            loss = lm_cross_entropy_loss(logits, tokens)
            total_loss += loss.item() * tokens.numel()
            total_tokens += tokens.numel()

    avg_loss = total_loss / total_tokens
    perplexity = math.exp(avg_loss)
    return perplexity
````

````python
def generate_text(model, tokenizer, prompt, max_length=100, temperature=1.0, top_k=50, device='cuda'):
    model.eval()
    tokens = tokenizer.encode(prompt, return_tensors='pt').to(device)

    for _ in range(max_length):
        # Get predictions
        with torch.no_grad():
            logits = model(tokens)

        # Get last token logits
        next_token_logits = logits[0, -1, :] / temperature

        # Apply top-k filtering
        if top_k is not None:
            indices_to_remove = next_token_logits < torch.topk(next_token_logits, top_k)[0][..., -1, None]
            next_token_logits[indices_to_remove] = -float('Inf')

        # Sample from distribution
        probs = F.softmax(next_token_logits, dim=-1)
        next_token = torch.multinomial(probs, num_samples=1)

        # Append to sequence
        tokens = torch.cat([tokens, next_token.unsqueeze(0)], dim=1)

        # Stop if we generate end-of-text token
        if next_token.item() == tokenizer.eos_token_id:
            break

    return tokenizer.decode(tokens[0], skip_special_tokens=True)
````

````python
class PosEmbed(nn.Module):
    def forward(self, tokens):
        # tokens: [batch, position]
        position = torch.arange(tokens.size(1), device=tokens.device)
        # Get positional embeddings [position, d_model] and expand to [batch, position, d_model]
        pos_embeddings = self.W_pos[position]  # [position, d_model]
        return pos_embeddings.unsqueeze(0).expand(tokens.size(0), -1, -1)  # [batch, position, d_model]

class MLP(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.W_in = nn.Parameter(torch.empty((cfg.d_model, cfg.d_mlp)))
        nn.init.normal_(self.W_in, std=self.cfg.init_range)
        self.b_in = nn.Parameter(torch.zeros((cfg.d_mlp)))
        self.W_out = nn.Parameter(torch.empty((cfg.d_mlp, cfg.d_model)))
        nn.init.normal_(self.W_out, std=self.cfg.init_range)
        self.b_out = nn.Parameter(torch.zeros((cfg.d_model)))

    def forward(self, normalized_resid_mid):
        # normalized_resid_mid: [batch, position, d_model]
        # Add print statement to check input shape to MLP
        print(f"Input shape to MLP (normalized_resid_mid): {normalized_resid_mid.shape}")
        # Step 1: Linear transformation with W_in and b_in
        # YOUR CODE HERE - IMPLEMENT THE MLP FORWARD PASS
        # Perform a linear transformation: normalized_resid_mid @ W_in + b_in
        # Use einsum for clarity and broadcasting the bias
        hidden = einops.einsum(normalized_resid_mid, self.W_in, 'batch position d_model, d_model d_mlp -> batch position d_mlp') + self.b_in
        # Step 2: Apply GELU activation
        activated = gelu_new(hidden)
        # Step 3: Linear transformation with W_out and b_out
        output = einops.einsum(activated, self.W_out, 'batch position d_mlp, d_mlp d_model -> batch position d_model') + self.b_out
        return output

class TransformerBlock(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        # Initialize layers for attention and MLP as per your configuration
        self.ln1 = LayerNorm(cfg)
        self.attn = Attention(cfg)
        self.ln2 = LayerNorm(cfg)
        self.mlp = MLP(cfg)

    def forward(self, resid_pre):
        input_shape = resid_pre.shape

        # Attention block
        normalized_resid_pre = self.ln1(resid_pre)
        attn_output = self.attn(normalized_resid_pre)
        resid_mid = attn_output + resid_pre
        assert resid_mid.shape == input_shape, \
            f"Attention block changed shape from {input_shape} to {resid_mid.shape}"

        # MLP block
        normalized_resid_mid = self.ln2(resid_mid)
        print(f"Shape before MLP: {normalized_resid_mid.shape}")  # Debug print
        mlp_output = self.mlp(normalized_resid_mid)
        resid_post = mlp_output + resid_mid
        assert resid_post.shape == input_shape, \
            f"MLP block changed shape from {input_shape} to {resid_post.shape}"

        return resid_post

class DemoTransformer(nn.Module):
    def forward(self, tokens):
        batch_size, seq_len = tokens.shape

        # Token and positional embeddings
        embedded_tokens = self.embed(tokens)  # [batch, seq_len, d_model]
        pos_embedded = self.pos_embed(tokens)  # [batch, seq_len, d_model]

        # Verify embedding shapes
        assert embedded_tokens.shape == (batch_size, seq_len, self.cfg.d_model)
        assert pos_embedded.shape == (batch_size, seq_len, self.cfg.d_model)

        # Initial residual stream
        residual = embedded_tokens + pos_embedded

        # Pass through transformer blocks
        for block in self.blocks:
            residual = block(residual)
            assert residual.shape == (batch_size, seq_len, self.cfg.d_model)

        # Final output
        normalized_residual = self.ln_final(residual)
        logits = self.unembed(normalized_residual)
        assert logits.shape == (batch_size, seq_len, self.cfg.d_vocab)

        return logits
````

````python
# Update test functions
def rand_float_test(cls, shape):
    cfg = Config(debug=True)
    # Separate instantiation and moving to GPU
    layer = cls(cfg)
    layer.cuda()
    random_input = torch.randn(shape).cuda() # Ensure input is on CUDA
    print("Input shape:", random_input.shape)
    output = layer(random_input)
    print("Output shape:", output.shape)
    print()
    return output

def rand_int_test(cls, shape):
    cfg = Config(debug=True)
    # Separate instantiation and moving to GPU
    layer = cls(cfg)
    layer.cuda()
    random_input = torch.randint(100, 1000, shape).cuda() # Ensure input is on CUDA
    print("Input shape:", random_input.shape)
    output = layer(random_input)
    print("Output shape:", output.shape)
    print()
    return output

def load_gpt2_test(cls, gpt2_layer, input_name, cache_dict=cache.cache_dict):
    cfg = Config(debug=True)
    # Separate instantiation and moving to GPU
    layer = cls(cfg)
    layer.cuda()
    layer.load_state_dict(gpt2_layer.state_dict(), strict=False)
    # Allow inputs of strings or tensors
    if isinstance(input_name, str):
        reference_input = cache_dict[input_name].cuda() # Ensure input is on CUDA
    else:
        reference_input = input_name.cuda() # Ensure input is on CUDA
    print("Input shape:", reference_input.shape)
    output = layer(reference_input)
    print("Output shape:", output.shape)
    reference_output = gpt2_layer(reference_input) # Ensure reference model also gets CUDA input
    print("Reference output shape:", reference_output.shape)

    comparison = torch.isclose(output, reference_output, atol=1e-4, rtol=1e-3)
    print(f"{comparison.sum()/comparison.numel():.2%} of the values are correct")
    return output

# Update PosEmbed (make sure this definition is run)
class PosEmbed(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        # W_pos is a lookup table for positional embeddings
        self.W_pos = nn.Parameter(torch.empty((cfg.n_ctx, cfg.d_model)))
        nn.init.normal_(self.W_pos, std=self.cfg.init_range)

    def forward(self, tokens):
        # tokens: [batch, position]
        # Create a tensor of positions [0, 1, 2, ..., position-1]
        position = torch.arange(tokens.size(1), device=tokens.device)
        # Lookup the positional embedding for each position in the batch
        # W_pos has shape [n_ctx, d_model]
        # We select the rows corresponding to the positions in the input sequence
        # Return with an added batch dimension for explicit broadcasting
        # And expand to the batch size of the input tokens
        pos_embeddings = self.W_pos[position] # [position, d_model]
        return pos_embeddings.unsqueeze(0).expand(tokens.size(0), -1, -1) # [batch, position, d_model]


# Update DemoTransformer (make sure this definition is run)
class DemoTransformer(nn.Module):
    def __init__(self, cfg):
        super().__init__()
        self.cfg = cfg
        self.embed = Embed(cfg)
        self.pos_embed = PosEmbed(cfg)
        self.blocks = nn.ModuleList([TransformerBlock(cfg) for _ in range(cfg.n_layers)])
        self.ln_final = LayerNorm(cfg)
        self.unembed = Unembed(cfg)

    def forward(self, tokens):
        # tokens [batch, position]
        # Step 1: Get token embeddings
        embedded_tokens = self.embed(tokens)
        # Step 2: Get positional embeddings
        # PosEmbed now returns [batch, position, d_model]
        pos_embedded = self.pos_embed(tokens)
        # Step 3: Combine embeddings (residual stream initially)
        # pos_embedded is now [batch, position, d_model], directly add to embedded_tokens
        residual = embedded_tokens + pos_embedded
        # Step 4: Pass through transformer blocks
        for block in self.blocks:
            residual = block(residual)
        # Step 5: Apply final layer norm
        normalized_residual = self.ln_final(residual)
        # Step 6: Apply unembedding
        logits = self.unembed(normalized_residual)

        return logits # Return the calculated logits


# Rerun the class definitions for LayerNorm, Attention, MLP, TransformerBlock, Unembed
# to ensure the most recent versions are in the environment, then run the test cell below.

# Verification Cell for Transformer Implementation (this cell)
# Run this cell AFTER running the updated class definitions above

print("="*50)
print("Testing Model Components")
print("="*50)

# Test LayerNorm
print("\nTesting LayerNorm:")
_ = rand_float_test(LayerNorm, [2, 4, 768])
_ = load_gpt2_test(LayerNorm, reference_gpt2.ln_final, "blocks.11.hook_resid_post")

# Test Embedding
print("\nTesting Embed:")
_ = rand_int_test(Embed, [2, 4])
# Ensure tokens is on CUDA for load_gpt2_test
tokens_cuda = tokens.cuda()
_ = load_gpt2_test(Embed, reference_gpt2.embed, tokens_cuda)

# Test Attention
print("\nTesting Attention:")
_ = rand_float_test(Attention, [2, 4, 768])
_ = load_gpt2_test(Attention, reference_gpt2.blocks[0].attn, cache["blocks.0.ln1.hook_normalized"])

# Test MLP
print("\nTesting MLP:")
_ = rand_float_test(MLP, [2, 4, 768])
_ = load_gpt2_test(MLP, reference_gpt2.blocks[0].mlp, cache["blocks.0.ln2.hook_normalized"])

# Test Transformer Block
print("\nTesting TransformerBlock:")
_ = rand_float_test(TransformerBlock, [2, 4, 768])
_ = load_gpt2_test(TransformerBlock, reference_gpt2.blocks[0], cache["resid_pre", 0])

# Test Unembedding
print("\nTesting Unembed:")
_ = rand_float_test(Unembed, [2, 4, 768])
_ = load_gpt2_test(Unembed, reference_gpt2.unembed, cache["ln_final.hook_normalized"])

# Test Full Model
print("\nTesting Full Model:")
_ = rand_int_test(DemoTransformer, [2, 4])
# Ensure tokens is on CUDA for load_gpt2_test
tokens_cuda = tokens.cuda()
_ = load_gpt2_test(DemoTransformer, reference_gpt2, tokens_cuda)


print("\n" + "="*50)
print("Qualitative Generation Test")
print("="*50)

# Test generation (Ensure demo_gpt2 is correctly initialized after the model definitions)
# Re-initialize and load state dict for demo_gpt2 after updating class definitions
demo_gpt2 = DemoTransformer(Config(debug=False))
demo_gpt2.load_state_dict(reference_gpt2.state_dict(), strict=False)
demo_gpt2.cuda()

test_prompt = "Once upon a time"
generated_text = test_prompt
for i in range(50):
    test_tokens = reference_gpt2.to_tokens(generated_text).cuda()
    # Use demo_gpt2 for generation
    demo_logits = demo_gpt2(test_tokens)
    generated_text += reference_gpt2.tokenizer.decode(demo_logits[0, -1].argmax()) # Corrected indexing

print("Generated Text:")
print(generated_text)

print("\n" + "="*50)
print("Loss Calculation Test")
print("="*50)

# Test loss calculation
# Ensure demo_gpt2 is correctly initialized after the model definitions
# (already done above for generation)
test_string = "This is a test string to calculate loss"
test_tokens = reference_gpt2.to_tokens(test_string).cuda()
demo_logits = demo_gpt2(test_tokens)
loss = lm_cross_entropy_loss(demo_logits, test_tokens)
print(f"Loss: {loss.item():.4f}")
print(f"Perplexity: {math.exp(loss.item()):.4f}")

print("\n" + "="*50)
print("Training Setup Verification")
print("="*50)

# Verify training setup
# Ensure 'model' is correctly initialized after the model definitions
model_cfg_for_training = Config(debug=False, d_model=256, n_heads=4, d_head=64, d_mlp=1024, n_layers=2, n_ctx=256, d_vocab=reference_gpt2.cfg.d_vocab)
model = DemoTransformer(model_cfg_for_training)
model = model.cuda()
print("Model architecture:")
print(model)
print("\nModel parameters:")
total_params = sum(p.numel() for p in model.parameters())
print(f"Total parameters: {total_params:,}")

# Verify data loader
sample_batch = next(iter(data_loader))
print("\nSample batch shapes:")
print(f"Tokens shape: {sample_batch['tokens'].shape}")

print("\nAll tests completed!")
````
